```python
import ffmpeg

def get_video_duration(file_path):
    probe = ffmpeg.probe(file_path)
    duration = float(probe['format']['duration'])
    return duration

video_file = '/home/thiendc/projects/video_summarization/sample_tv360.mp4'  # Thay thế bằng tên tệp của bạn
video_file_highlight = '/home/thiendc/projects/video_summarization/sample_tv360_highlight.mp4'
duration = get_video_duration(video_file)/get_video_duration(video_file_highlight)
print(f"Độ dài video: {duration} giây")
```

```python
# dài 12p x2 = 24 cảnh
def get_video_fps(file_path):
    probe = ffmpeg.probe(file_path)
    fps_info = probe['streams'][0]['r_frame_rate']
    numerator, denominator = map(int, fps_info.split('/'))
    fps = numerator / denominator
    return fps

video_file = '/home/thiendc/projects/video_summarization/sample_tv360_highlight.mp4'  # Thay thế bằng tên tệp của bạn
fps = get_video_fps(video_file)
print(f"FPS của video: {fps}")
```

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
from sklearn.metrics.pairwise import cosine_similarity

def extract_keyframes(video_path, interval=30):
    cap = cv2.VideoCapture(video_path)
    keyframes = []
    frame_count = 0
    success, frame = cap.read()
    
    while success:
        if frame_count % interval == 0:  # Trích xuất mỗi 'interval' frame
            keyframes.append(frame)
        success, frame = cap.read()
        frame_count += 1

    cap.release()
    return keyframes

extract_keyframes("/home/thiendc/projects/video_summarization/data/sample_tv360/gt/chunk_1_gt.mp4")
```
