```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# ng dn n nh v nhn
image_path = '/home/thiendc/projects/video_summarization/v2/data/scorebox_detection/train/images/BPL_22_1_high_frame_5000_jpg.rf.6e197b1db9542dae3defe32b67a865a8.jpg'
label_path = '/home/thiendc/projects/video_summarization/v2/data/scorebox_detection/train/labels/BPL_22_1_high_frame_5000_jpg.rf.6e197b1db9542dae3defe32b67a865a8.txt'

# c nh
image = Image.open(image_path)
width, height = image.size

# c v x l tp nhn
with open(label_path, 'r') as f:
    lines = f.readlines()

# To hnh nh v v cc bounding box
fig, ax = plt.subplots(1)
ax.imshow(image)

for line in lines:
    # Tch cc gi tr t dng (format: class x_center y_center width height)
    label, x_center, y_center, box_width, box_height = map(float, line.strip().split())

    # Chuyn i t chun ha sang ta  pixel
    x_center *= width
    y_center *= height
    box_width *= width
    box_height *= height

    # Tnh ta  gc trn bn tri ca bounding box
    x_min = x_center - box_width / 2
    y_min = y_center - box_height / 2

    # To v thm bounding box vo nh
    rect = patches.Rectangle((x_min, y_min), box_width, box_height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    # Gn nhn vo bounding box
    ax.text(x_min, y_min, f'Class: {int(label)}', color='white', fontsize=8, bbox=dict(facecolor='red', alpha=0.5))

# Hin th nh vi bounding box
plt.axis('off')
plt.show()
```

```python
import os
from tqdm import tqdm
label_dirs = [f'./data/scorebox_detection/{split}/labels/' for split in ['train', 'valid', 'test']]
for label_dir in tqdm(label_dirs):
    for filename in tqdm(os.listdir(label_dir)):
        if filename.endswith('.txt'):
            label_path = os.path.join(label_dir, filename)
            with open(label_path, 'r') as file:
                lines = file.readlines()
                
            # Lc b cc dng c label = 1
            filtered_lines = [line for line in lines if not line.startswith('1 ')]

            # Ghi  li tp vi cc dng  lc
            with open(label_path, 'w') as file:
                file.writelines(filtered_lines)
                print(f" xa cc dng c label = 1 trong {label_path.split('/')[-1]}")
            
```

```python
import os
from tqdm import tqdm

# ng dn ti cc th mc nhn
label_dirs = [f'./data/scorebox_detection/{split}/labels/' for split in ['train', 'valid', 'test']]

# Duyt qua tng th mc nhn
for label_dir in tqdm(label_dirs, desc="Processing label directories"):
    for filename in tqdm(os.listdir(label_dir), desc=f"Processing files in {label_dir}"):
        if filename.endswith('.txt'):  # Ch x l cc tp .txt
            label_path = os.path.join(label_dir, filename)
            
            # c ni dung ca tp nhn
            with open(label_path, 'r') as file:
                lines = file.readlines()
                
            # Thay i cc dng c nhn bt u bng '2' thnh '1'
            modified_lines = [line.replace('2 ', '1 ', 1) if line.startswith('2 ') else line for line in lines]

            # Ghi  li tp vi cc dng  c thay i
            with open(label_path, 'w') as file:
                file.writelines(modified_lines)
            
            # Thng bo khi hon tt x l tp
            print(f" thay i nhn '2' thnh '1' trong {filename}")

print("Hon tt x l tt c cc tp nhn.")
```

```python
import time
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2

# # Bc 1: Tnh thi gian ti m hnh
# start_time = time.time()
# model = YOLO("runs/train/scoreboard_detection8/weights/last.pt")
# load_model_time = time.time() - start_time
# print(f"Thi gian ti m hnh: {load_model_time:.4f} giy")

# Bc 2: Tnh thi gian inference
start_time = time.time()
results = model("/home/thiendc/projects/video_summarization/v2/data/sample/ocr1.png")
inference_time = time.time() - start_time
print(f"Thi gian inference: {inference_time:.4f} giy")

# Bc 3: Tnh thi gian c nh
start_time = time.time()
image_path = "/home/thiendc/projects/video_summarization/v2/data/sample/ocr1.png"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Chuyn sang RGB  ph hp vi matplotlib
read_image_time = time.time() - start_time
print(f"Thi gian c nh: {read_image_time:.4f} giy")

# Bc 4: Hin th bounding boxes v tnh thi gian hin th kt qu
start_time = time.time()
for result in results:
    boxes = result.boxes
    for box in boxes:
        cls = int(box.cls)  # Lp d on
        conf = float(box.conf)  #  chnh xc
        x1, y1, x2, y2 = box.xyxy[0].tolist()  # Ta  bounding box

        # V bounding box ln nh
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
        cv2.putText(image, f"{cls}: {conf:.2f}", (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Hin th nh vi bounding boxes
plt.imshow(image)
plt.axis("off")
plt.show()
display_time = time.time() - start_time
print(f"Thi gian hin th kt qu: {display_time:.4f} giy")
```
