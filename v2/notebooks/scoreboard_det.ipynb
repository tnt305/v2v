```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Đường dẫn đến ảnh và nhãn
image_path = '/home/thiendc/projects/video_summarization/v2/data/scorebox_detection/train/images/BPL_22_1_high_frame_5000_jpg.rf.6e197b1db9542dae3defe32b67a865a8.jpg'
label_path = '/home/thiendc/projects/video_summarization/v2/data/scorebox_detection/train/labels/BPL_22_1_high_frame_5000_jpg.rf.6e197b1db9542dae3defe32b67a865a8.txt'

# Đọc ảnh
image = Image.open(image_path)
width, height = image.size

# Đọc và xử lý tệp nhãn
with open(label_path, 'r') as f:
    lines = f.readlines()

# Tạo hình ảnh và vẽ các bounding box
fig, ax = plt.subplots(1)
ax.imshow(image)

for line in lines:
    # Tách các giá trị từ dòng (format: class x_center y_center width height)
    label, x_center, y_center, box_width, box_height = map(float, line.strip().split())

    # Chuyển đổi từ chuẩn hóa sang tọa độ pixel
    x_center *= width
    y_center *= height
    box_width *= width
    box_height *= height

    # Tính tọa độ góc trên bên trái của bounding box
    x_min = x_center - box_width / 2
    y_min = y_center - box_height / 2

    # Tạo và thêm bounding box vào ảnh
    rect = patches.Rectangle((x_min, y_min), box_width, box_height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    # Gán nhãn vào bounding box
    ax.text(x_min, y_min, f'Class: {int(label)}', color='white', fontsize=8, bbox=dict(facecolor='red', alpha=0.5))

# Hiển thị ảnh với bounding box
plt.axis('off')
plt.show()
```

```python
import os
from tqdm import tqdm

# Đường dẫn tới các thư mục nhãn
label_dirs = [f'./data/scorebox_detection/{split}/labels/' for split in ['train', 'valid', 'test']]

# Duyệt qua từng thư mục nhãn
for label_dir in tqdm(label_dirs, desc="Processing label directories"):
    for filename in tqdm(os.listdir(label_dir), desc=f"Processing files in {label_dir}"):
        if filename.endswith('.txt'):  # Chỉ xử lý các tệp .txt
            label_path = os.path.join(label_dir, filename)
            
            # Đọc nội dung của tệp nhãn
            with open(label_path, 'r') as file:
                lines = file.readlines()
                
            # Thay đổi các dòng có nhãn bắt đầu bằng '2' thành '1'
            modified_lines = [line.replace('2 ', '1 ', 1) if line.startswith('2 ') else line for line in lines]

            # Ghi đè lại tệp với các dòng đã được thay đổi
            with open(label_path, 'w') as file:
                file.writelines(modified_lines)
            
            # Thông báo khi hoàn tất xử lý tệp
            print(f"Đã thay đổi nhãn '2' thành '1' trong {filename}")

print("Hoàn tất xử lý tất cả các tệp nhãn.")
```

```python
from ultralytics import YOLO
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '2, 3'
# Tải mô hình YOLOv8s pre-trained
model = YOLO('yolov8s.pt')

# Fine-tune mô hình với dữ liệu của bạn
results = model.train(
    data='/home/thiendc/projects/video_summarization/v2/data/scorebox_detection/data.yaml',  # Đường dẫn tới file cấu hình data.yaml
    epochs=50,         # Số epoch để huấn luyện
    batch= 32,          # Batch size
    imgsz=640,         # Kích thước ảnh (mặc định là 640x640)
    project='runs/train',  # Thư mục để lưu các mô hình và log
    name='scoreboard_detection',        # Tên của lần chạy huấn luyện
    pretrained=True,   # Sử dụng trọng số từ mô hình YOLOv8s pre-trained,
    device = '2,3 ',
    dropout=  0.1,
)
```

```python
import time
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2

# # Bước 1: Tính thời gian tải mô hình
# start_time = time.time()
# model = YOLO("runs/train/scoreboard_detection8/weights/last.pt")
# load_model_time = time.time() - start_time
# print(f"Thời gian tải mô hình: {load_model_time:.4f} giây")

# Bước 2: Tính thời gian inference
start_time = time.time()
results = model("/home/thiendc/projects/video_summarization/v2/data/sample/ocr1.png")
inference_time = time.time() - start_time
print(f"Thời gian inference: {inference_time:.4f} giây")

# Bước 3: Tính thời gian đọc ảnh
start_time = time.time()
image_path = "/home/thiendc/projects/video_summarization/v2/data/sample/ocr1.png"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Chuyển sang RGB để phù hợp với matplotlib
read_image_time = time.time() - start_time
print(f"Thời gian đọc ảnh: {read_image_time:.4f} giây")

# Bước 4: Hiển thị bounding boxes và tính thời gian hiển thị kết quả
start_time = time.time()
for result in results:
    boxes = result.boxes
    for box in boxes:
        cls = int(box.cls)  # Lớp dự đoán
        conf = float(box.conf)  # Độ chính xác
        x1, y1, x2, y2 = box.xyxy[0].tolist()  # Tọa độ bounding box

        # Vẽ bounding box lên ảnh
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
        cv2.putText(image, f"{cls}: {conf:.2f}", (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Hiển thị ảnh với bounding boxes
plt.imshow(image)
plt.axis("off")
plt.show()
display_time = time.time() - start_time
print(f"Thời gian hiển thị kết quả: {display_time:.4f} giây")
```
