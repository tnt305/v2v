```python
import os
import subprocess
from tqdm import tqdm
def get_video_duration(video_path):
    result = subprocess.run(
        ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", video_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT
    )
    return float(result.stdout)

data_path = './data/sample_tv360/do_merge'
gt_path = './data/sample_tv360/gt'
data = sorted(os.listdir('./data/sample_tv360/do_merge'), key = lambda x: int(x.split("_")[-1].split(".")[0]))
gt = sorted(os.listdir('./data/sample_tv360/gt'), key = lambda x: int(x.split("_")[1].split(".")[0]))
for i, j in tqdm(zip(data, gt), total=len(data), desc="Processing videos"):
    duration1 = get_video_duration(f"{data_path}/{i}")
    duration2 = get_video_duration(f"{gt_path}/{j}")
    common_duration = min(duration1, duration2)
    
    subprocess.run(["ffmpeg", '-y',"-i", f"{data_path}/{i}", "-t", str(common_duration), "-c", "copy", f"{data_path}/{i}"])
    subprocess.run(["ffmpeg", '-y',"-i", f"{data_path}/{j}", "-t", str(common_duration), "-c", "copy", f"{data_path}/{j}"])
```

```python
import subprocess

# Đường dẫn đến video
video1_path = "./data/sample_tv360/final/complete_video.mp4"
video2_path = "./data/sample_tv360/final/complete_video_gt.mp4"
output1_path = "./data/sample_tv360/final/trimmed_complete_video.mp4"
output2_path = "./data/sample_tv360/final/trimmed_complete_video_gt.mp4"

# Lấy thời lượng của hai video
def get_video_duration(video_path):
    result = subprocess.run(
        ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", video_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT
    )
    return float(result.stdout)

# Tính thời gian chung
duration1 = get_video_duration(video1_path)
duration2 = get_video_duration(video2_path)
common_duration = min(duration1, duration2)

# Cắt hai video sao cho có cùng độ dài
subprocess.run(["ffmpeg", "-i", video1_path, "-t", str(common_duration), "-c", "copy", output1_path])
subprocess.run(["ffmpeg", "-i", video2_path, "-t", str(common_duration), "-c", "copy", output2_path])
```
